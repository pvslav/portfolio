```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "booking_id": [1001, 1002, 1003, 1004, 1005],
    "property" : ["A", "A", "B", "B", "C"],
    "created_at": ["2022-03-01", "2022-02-10", "2022-04-12",
                   "2022-04-11", "2022-06-05"],
    "checkin_date": ["2022-06-01", "2022-06-10", "2022-06-02",
                     "2022-06-20", "2022-08-10"],
    "checkout_date": ["2022-06-06", "2022-06-15", 
                      "2022-06-06","2022-06-28", "2022-08-16"],
    "amount": [5400, 5600, 4800, 9000, 6500]
})
# Change type of data
date_cols = ["created_at","checkin_date","checkout_date"]
df[date_cols] = df[date_cols].astype("datetime64[ns]")
# Output of the Dataframe
df
```

#### How to extract information about the month and year from the date?

```python
# extracting the month and day
month = df['created_at'].dt.month[0]
day = df['created_at'].dt.day[0]
print(month,day)
```

```python
# Creating a column of year_month
df["year_month"] = df["created_at"].dt.to_period("M")
# Output of the Dataframe
df
```

#### How to add a time interval to a date?

```python
df.loc[df["booking_id"]==1002, "checkout_date"] = \
df.loc[df["booking_id"]==1002, "checkout_date"] + \
pd.DateOffset(days=1)
print(df.loc[df["booking_id"]==1002, "checkout_date"])
```

#### How to find an interval in days between two dates?

```python
# The difference expressed in days
df["days_to_checkin"] = \
(df["checkin_date"] - df["created_at"]).dt.days
# Output of DataFrame
df
```
